```python
import matplotlib.pyplot as plt
import plotnine as p9
import numpy as np
import pandas as pd

# Prepare plotting:
# "matplotlib inline" is not necessary as set by default
%matplotlib inline
plt.rcParams['figure.figsize'] = [15, 10] # figure size

# %pylab inline
# pylab.rcParams['figure.figsize'] = [15, 10] # figure size
```

http://pythonplot.com/#simple-histogram

```python
theta = 3
attachment_point = 2
inflation_rate = np.linspace(attachment_point, 5, 1000)
tail_inflation = theta * np.exp(-theta * inflation_rate)

reserve_relief_rate = 0.05 / 0.95 * np.mean(inflation_rate)
reserve_relief = [-reserve_relief_rate, 0]# np.repeat(reserve_relief_rate, 1)
```

```python
plt.plot(inflation_rate, tail_inflation)
np.mean(inflation_rate)
```

```python

default_color = 1

tail_df = pd.DataFrame(tail_inflation, columns=['PDF'])
tail_df['Inflation_Rate'] = inflation_rate
tail_df['Transparency'] = 1
tail_df['Color'] = default_color
# tail_df.transparancy = tail_df.transparancy.astype('category')

def transparency(x):
    if x == 0: return 0
    else: return 1
    
def color(x):
    if x == 0: return 0
    else: return default_color

relief_df = pd.DataFrame(reserve_relief, columns = ['Inflation_Rate'])
relief_df['PDF'] = 0.1
relief_df['Transparency'] = relief_df['Inflation_Rate'].apply(transparency)
relief_df['Color'] = relief_df['Inflation_Rate'].apply(color)
#relief_df['transparency'] = pd.Categorical(relief_df['Inflation_Rate'], categories=[0, 0.3])
#relief_df['transparency'] = relief_df['transparency'].astype(float) 


# gg = p9.ggplot(relief_df) + p9.geom_histogram(binwidth=0.05)
gg = p9.ggplot() + p9.aes(x='Inflation_Rate', y='PDF', alpha='Color', colour='Color') # colour='Color', fill='Color', alpha='Transparency', fill='transparency'
gg = gg + p9.geom_line(data=tail_df) # , colour='yellow'
gg = gg + p9.geom_area(data=tail_df) # , alpha=0.3
gg = gg + p9.geom_bar(data = relief_df, stat = "identity", width=0.1)

# gg = gg + p9.ylim(0, 0.1) 
# gg = gg + p9.expand_limits(y = 0.05)
# gg = gg + p9.scale_y_continuous(expand = (0.01,0))
# gg = gg + p9.coord_cartesian(ylim = [0, 0.03])

# gg = gg + p9.scale_fill_gradient(low = "white", high = "green")
gg = gg + p9.scale_color_gradient(low='#E1FA72', high='#F46FEE')
gg = gg + p9.guides(alpha=False, fill=False, color=False) # do not show legend for alpha (transparency)

# p9.scale_color_brewer(type='div', palette=7) # http://docs.ggplot2.org/current/scale_brewer.html
# scale_fill_gradient("Count", low = "blue", high = "red")
#  scale_fill_distiller(palette = "Spectral")
# aes(fill=..count..)
# https://ggplot2.tidyverse.org/reference/scale_gradient.html

# Save plot
gg.save("scatterplot.png", width=10, height=10, dpi=300)

gg
```
