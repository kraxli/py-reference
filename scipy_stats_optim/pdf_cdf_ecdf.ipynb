```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.ticker import FormatStrFormatter, ScalarFormatter
```

```python
def ecdf(a):
    x, counts = np.unique(a, return_counts=True)
    cusum = np.cumsum(counts)
    ecdf = cusum / cusum[-1]
    return x, ecdf

size = int(10e3)
amounts = np.exp(np.random.randn(size))
amounts, ecdf = ecdf(amounts)


## Plot:
plt.style.use('ggplot')
# data = pd.DataFrame({'AMOUNT': np.log(amounts), 'ECDF': ecdf})
data = pd.DataFrame({'AMOUNT': amounts, 'ECDF': ecdf})

ax = data.plot.scatter(x='AMOUNT', y='ECDF', logx=True, color='lightblue', s=0.1)

data.set_index('AMOUNT', inplace=True)
ax = data.plot(logx=True, color='lightblue')  # logx=True, loglog=True

ax.yaxis.set_major_formatter(ScalarFormatter(useOffset=True))
ax.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
# ax.set_yticks(np.arange(0.1, 1.1, 0.1))

# Note: you can usually get the current figure and axes objects using the following functions

# fig = plt.gcf()
# ax = plt.gca()
# fig,ax=plt.subplots(5,4)
# ax_list = fig.axes
```

```python
import scipy
import matplotlib.pyplot as plt
import seaborn as sns

x = np.random.randn(10000) # generate samples from normal distribution (discrete data)
norm_cdf = scipy.stats.norm.cdf(x) # calculate the cdf - also discrete

# plot the cdf
sns.lineplot(x=x, y=norm_cdf)
plt.show()
```
